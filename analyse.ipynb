On remarque que :
"Single" , "Alone" sont pareil
"Together" , "Married" aussi son pareil
"Absurd" et "YOLO" sont aberantes


```python
data_company["Marital_Status"] = data_company["Marital_Status"].replace(
    ["Alone", "window"], "Single")

data_company["Marital_Status"] = data_company["Marital_Status"].replace(
    "Together", "Married")
```

```python
data_company
```

```python
# Filtrer les clients avec Marital_Status "Absurd" ou "YOLO"
clients_filtrÃ©s = data_company[(data_company["Marital_Status"] == "Absurd") | (
    data_company["Marital_Status"] == "YOLO")]

# Afficher la liste des clients filtrÃ©s
clients_filtrÃ©s
```

```python
data_company = data_company[~((data_company["Marital_Status"] == "Absurd") | (
    data_company["Marital_Status"] == "YOLO"))]
data_company
```

```python
# numerison les valeurs de Education
# Transformation des colonnes 'Education' et 'Marital_Status' en dummies
data_company_dummies = pd.get_dummies(
    data_company, columns=["Education", "Marital_Status"], prefix=["Edu", "Marital"])

data_company_dummies
final_data_company = data_company_dummies
```

```python
sns. countplot(data=data_company, x="Response")
```

```python
from datetime import datetime
```

```python
final_data_company['Dt_Customer'] = (
    datetime.now() - final_data_company['Dt_Customer']).dt.days
final_data_company
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X = final_data_company.drop('Response', axis=1)
y = final_data_company['Response']
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=101)
```

```python
scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

## Utilisons le model de regression Logistique


```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
```

```python
log_model = LogisticRegression()
```

```python
log_model.fit(scaled_X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, ConfusionMatrixDisplay
```

```python
y_pred = log_model.predict(scaled_X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
